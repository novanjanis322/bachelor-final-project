```python
sentiment_df = pd.read_excel('./drive/MyDrive/TA Tugas Akhir Workspace/Dataset/Train/labelled_data.xlsx')
sentiment_df.head(3)
```

```python
sentiment_map = dict(zip(sentiment_df['text'], sentiment_df['sentiment']))
```

```python
directory = './drive/MyDrive/TA Tugas Akhir Workspace/Dataset/Train'
csv_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.csv')]
csv_files
```

```python
for file in csv_files:
    df = pd.read_csv(file)
    if 'Text' in df.columns:
      df['sentiment'] = df['Text'].map(sentiment_map)
    elif 'textOriginal' in df.columns:
      df['sentiment'] = df['textOriginal'].map(sentiment_map)
    # Save the updated DataFrame to a new CSV file
    output_file = os.path.join(directory, 'updated_' + os.path.basename(file))
    df.to_csv(output_file, index=False)
```
